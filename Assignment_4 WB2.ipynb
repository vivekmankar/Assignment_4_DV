```python
import pandas as pd 
Location = "datasets/gradedata.csv" 
df = pd.read_csv(Location) 
df.head()
```

```python
import numpy as np 
df['isFailing'] = np.where(df['grade']<70, 'yes', 'no') 
df.tail(10)
```

```python
#Can you create a column for timemgmt that shows busy if a student exercises more than three hours per week 
#AND studies more than seventeen hours per week
```
