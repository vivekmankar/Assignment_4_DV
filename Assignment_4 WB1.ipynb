```python
import pandas as pd 
Location = "datasets/gradedata.csv" 
df = pd.read_csv(Location) 
df.head()
```

```python
bins = [0,70,100]
```

```python
group_names = ['Fail','Pass']
```

```python
df['Results'] = pd.cut(df['grade'], bins,           labels=group_names)
```

```python
#Import the datasets/gradedata.csv file and create a new binned column of the 'status' as either passing (> 70) or failing (<=70). 
#Then, compute the mean hours of exercise of the female students with a 'status' of passing.
```

```python
df[df['Results']=='Pass'].groupby('gender').agg( {"hours":"mean"} )
```
